```python
from skimage.color import rgb2gray,bgr2rgb
from skimage import data
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.io import imread
```

```python
img = data.astronaut()
img = rgb2gray(img)
low_sigma1 = cv2.GaussianBlur(img,(31,31),1)
low_sigma2 = cv2.GaussianBlur(img,(31,31),2)
low_sigma3 = cv2.GaussianBlur(img,(31,31),3)
low_sigma4 = cv2.GaussianBlur(img,(31,31),4)
high_sigma = cv2.GaussianBlur(img,(31,31),7)
```

```python
dog1=low_sigma1-high_sigma
dog2=low_sigma2-high_sigma
dog3=low_sigma3-high_sigma
dog4=low_sigma4-high_sigma
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
plt.title('Original Image')
plt.subplot(2,2,2)
plt.imshow(low_sigma1, cmap='gray')
plt.title('Blur with variance=1')
plt.subplot(2,2,3)
plt.imshow(low_sigma2, cmap='gray')
plt.title('Blur with variance=2')
plt.subplot(2,2,4)
plt.imshow(low_sigma3, cmap='gray')
plt.title('Blur with variance=3')


```

```python
mx=np.max(dog1)
th=mx*0.2
dog1[np.abs(dog1)>th]=255
dog2[np.abs(dog2)>th]=255
dog3[np.abs(dog3)>th]=255
dog4[np.abs(dog4)>th]=255
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(dog1,cmap='gray')
plt.title('large difference in variance')
plt.subplot(2,2,2)
plt.imshow(dog2, cmap='gray')
plt.title('Medium difference in variance')
plt.subplot(2,2,3)
plt.imshow(dog3, cmap='gray')
plt.title('Small diffference in variance')
plt.subplot(2,2,4)
plt.imshow(dog4, cmap='gray')
plt.title('Very Small diffference in variance')


```
