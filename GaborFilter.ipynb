```python
angle=(np.pi)/2
sigma=5
k_size=25
lamda=20
gamma=0.5
psi=0
```

```python
image = cv2.imread('/content/hidden_image.JPG',cv2.IMREAD_GRAYSCALE)
```

```python
filter = cv2.getGaborKernel((k_size, k_size), sigma, angle, lamda, gamma, psi)
```

```python
filter = filter/filter.sum()
```

```python
image_filtered = cv2.filter2D(image, -1, filter)
```

```python
plt.imshow(image_filtered,cmap='gray')
```

```python
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.subplot(1,2,2)
plt.imshow(image_filtered,cmap='gray')
```

***Conclusion:***


Gabor filter of size 101x101 is constructed for the following parameter values:
sigma=5 angle=0 lamda=10 gamma=0.5 psi=0

If sigma is increased from 5 to 10, gabor filter has more number of stripes which signifies that it can identify several edges.

Angle is changed from 0 to 90degrees which rotates gabor filter from vertical to horizontal axis which signifies that gabor filter can be used to identify edges in many directions.

When wavelength(lamda) is increased from 10 to 20, number of stripes of gabor filter reduces.

Gamma (ellipticity) controls identification of length and width of the edge. For gamma=0, entire edge is detected. If gamma is increased to 0.5 and 0.9, length of detected edge reduces and width increases.
